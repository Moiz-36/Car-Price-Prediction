```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
```

```python
dataset=pd.read_csv('car data.csv')
```

```python
dataset.head()
```

```python
print(dataset.Fuel_Type.value_counts())
print(dataset.Seller_Type	.value_counts())
print(dataset.Transmission.value_counts())
```

```python
dataset.replace({'Fuel_Type':{'Petrol':0,'Diesel':1,'CNG':2},'Seller_Type':{'Dealer':0,'Individual':1 },'Transmission':{'Manual':0,'Automatic':1}},inplace=True)
```

```python
dataset.head()
```

```python
x=dataset.drop(['Car_Name','Selling_Price'],axis=1)
y=dataset['Selling_Price']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.1,random_state=2)
```

linear regression

```python
lin_reg=LinearRegression()
```

```python
lin_reg.fit(xtrain,ytrain)
```

```python
tarining_data_prediction=lin_reg.predict(xtrain)
```

```python
error_score=metrics.r2_score(ytrain,tarining_data_prediction)
print(error_score)
```

```python
plt.scatter(ytrain,tarining_data_prediction)
plt.xlabel("actual price")
plt.ylabel("predicted price")
plt.title("actual price vs predicted price")
plt.show()
```

```python
testing_data_prediction=lin_reg.predict(xtest)
```

```python
error_score=metrics.r2_score(ytest,testing_data_prediction)
print(error_score)
```

```python
plt.scatter(ytest,testing_data_prediction)
plt.xlabel("actual price")
plt.ylabel("predicted price")
plt.title("actual price vs predicted price")
plt.show()
```
